```python
import pandas as pd
import numpy as np
```

```python
ads = pd.read_csv('ads_dataset.tsv',delimiter='\t',encoding='utf-8') #create dataframe
ads[:5]
```

```python
def getDfSummary(input_data):
    stats = pd.DataFrame(input_data.describe(include=[np.number]).transpose()) #stats but only include #s
    number_nan = pd.DataFrame(input_data.isna().sum(),columns = ['number_nan']) #NaN value count
    number_distinct = pd.DataFrame(input_data.nunique(),columns = ['number_distinct']) #unique values
    output_data = pd.concat([stats, number_nan, number_distinct], axis=1) #axis=1: stack to right of first col
    output_data = output_data.drop("count", axis = 1)
    
    return output_data

getDfSummary(ads)
```

```python
%timeit getDfSummary(ads)
```

```python
output_data = pd.DataFrame(getDfSummary(ads))
i=0
for x in output_data['number_nan']: # check if any rows have NaN values by count
    if x > 0:
        print(output_data.index[i])
    i+=1
    
```

There are a few fields that indicate and correlate perfectly to show that data is missing. These fields include 'isbuyer', 'buy_interval', 'expected_time_buy', and 'multiple_buy'. These fields now only show 1 distinct numerical value (of 0) for all rows of data, when previously they were binary ('isbuyer' and 'multiple_buy') or contained many distinct numbers ('buy_interval' and 'expected_time_buy').
